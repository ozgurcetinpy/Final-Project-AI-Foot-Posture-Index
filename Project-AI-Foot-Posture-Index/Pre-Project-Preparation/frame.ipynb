```python

from tensorflow.keras.models import load_model
import cv2
import numpy as np

model = load_model(r'C:\Users\ozgur\Desktop\Final Project\Bitirme Projesi\model\RMS00001_09_80_90.h5')

cap = cv2.VideoCapture(0)
while True:

    ret, frame = cap.read()
    
    resized = cv2.resize(frame, (150,150))
    normalized = resized / 255.0
    np_array = np.array(resized)
    frame = np.expand_dims(np_array, axis=0)
    result = model.predict(frame)
    print(result)
    confidence = result[0][0]
    if confidence <= 0.50: 
        label = "arka"
        _,height, width, channels = frame.shape
        x1, y1 = int(width * result[0][0]), int(height * result[0][0])
        x2, y2 = int(width * result[0][0]), int(height * result[0][0])
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
    else:
        label = "yan"
        _,height, width, channels, = frame.shape
        x1, y1 = int(width * result[0][0]), int(height * result[0][0])
        x2, y2 = int(width * result[0][0]), int(height * result[0][0])
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        
    cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow("Kamera", normalized)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('foot.jpg')
img = cv2.resize(img, (64, 64))
img = img.reshape(1, 64, 64, 3)
prediction = model.predict(img)
if prediction[0][0] > prediction[0][1]:
    print('Bu görüntüde ayak yandan görülüyor.')
else:
    print('Bu görüntüde ayak arkadan görülüyor.')
while True:
    # Görüntü yükleme
    ret,frame = cv2.VideoCapture(0)
    # Görüntüyü yeniden boyutlandırma ve yeniden şekillendirme
    img_resized = cv2.resize(frame, (64, 64))
    img_resized = img_resized.reshape(1, 64, 64, 3)
    # Tahmin yapma
    prediction = model.predict(img_resized)
    print(prediction)
    # Tahmin edilen ayağın konumunu belirleme
    foot_pos = np.argmax(prediction)
    # Kutu çizdirme
    if foot_pos == 0:
        cv2.rectangle(img, (10, 10), (50, 50), (0, 255, 0), 2)
    else:
        cv2.rectangle(img, (10, 10), (50, 50), (0, 0, 255), 2)
    # Görüntüyü gösterme
    cv2.imshow('Tahmin Edilen Ayak', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
